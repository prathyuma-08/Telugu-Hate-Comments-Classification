# Pre-process the transliterated column

```python
import pandas as pd
```

```python
test_new = pd.read_excel('test_transliterated.xlsx')
train_new = pd.read_excel('train_transliterated.xlsx')
```

```python
test_new.head()
```

```python
train_new.head()
```

```python
import re

def preprocess_text(text):
    # Remove symbols, special characters, and numbers
    text = re.sub(r'[^A-Za-z\s]', '', text)
    # Remove extra spaces and convert to lowercase
    text = ' '.join(text.split())
    text = text.lower()
    return text
```

```python
# Assuming 'test' is your DataFrame with a 'transliteration' column
test_new['transliterated'] = test_new['transliterated'].apply(preprocess_text)
```

```python
test_new.head()
```

```python
train_new['transliterated'] = train_new['transliterated'].apply(preprocess_text)
```

```python
train_new.head()
```

# Produce Word Tokens

```python
import nltk
from nltk.tokenize import word_tokenize

# Assuming 'test' is your DataFrame with a 'transliteration' column

# Download NLTK data (if you haven't already)
nltk.download('punkt')

# Tokenize the 'transliteration' column and store the results in a new column 'tokenized_transliteration'
test_new['tokenized_transliteration'] = test_new['transliterated'].apply(lambda text: word_tokenize(text))
```

```python
train_new['tokenized_transliteration'] = train_new['transliterated'].apply(lambda text: word_tokenize(text))
```
